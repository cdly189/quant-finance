### Quantnet Goldman Sach Suggested Reading List

```python
from bs4 import BeautifulSoup
import requests

source = requests.get('https://quantnet.com/threads/the-goldman-sachs-suggested-reading-list.4916/').text

soup = BeautifulSoup(source,'lxml')


for name in soup.find_all('a', class_='link link--external'):
    print(name.text)
```
